```python
import sympy
sympy.init_printing()

%matplotlib inline
```

```python
t = sympy.Symbol('t')
```

```python
sympy.plot(sympy.sin(t))
```

```python
x, y = sympy.symbols('x, y')
```

```python
f2 = sympy.sin(x) + sympy.cos(y)
```

```python
sympy.plotting.plot3d(f2)
```

```python
import numpy
import matplotlib.pyplot as plt
```

```python
f2numeric = sympy.lambdify((x, y), f2, 'numpy')
```

```python
xx, yy = numpy.mgrid[-10:10:0.1, -10:10:0.1]
```

```python
zz = f2numeric(xx, yy)
```

```python
plt.pcolor(xx, yy, zz)
```
